# 样本

```python
from datetime import datetime, timedelta
ds = pd.read_csv('../../yuChuli/parsedFlowTranspose.csv',index_col=0,parse_dates=True)
ds.columns = pd.to_datetime(ds.columns)
ds = ds[pd.date_range('2016-10-11','2016-10-31')]
# print (ds)

# 选中所有的礼拜二
# 样本构建
temp = np.zeros([3960,15])
count=0
tuesdayDates = []
for tuesday in ds.columns:
    for index in index10:
        # 选中礼拜二
        if tuesday.isoweekday() == 2:
            tmp = ds.loc[index,tuesday:tuesday + timedelta(13)]
#             print (len(tmp))
            panduan =  (len(tmp)>13)
            

            if  panduan: # 某行dataframe中的空值个数
                temp[count, 0] = index
                temp[count, 1:8] = tmp[tuesday:tuesday + timedelta(6)]
                tuesdayDates.append(tuesday)
                temp[count, 8:] = tmp[tuesday + timedelta(7) : tuesday + timedelta(13)]
#                 print (tmp[tuesday + timedelta(7) : tuesday + timedelta(13)])
                count = count +1
sampl =  pd.DataFrame(temp)

# print (sampl)
# print (count)

# sampl_x, sampl_y 分开
sampl_x = sampl.loc[:,1:7]
sampl_dates = pd.DataFrame({'date': tuesdayDates})
sampl_y = sampl.loc[:,8:]
```

```python
# 预处理 
# 排除异常：把NaN用当周的中位数 Imputer
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='median', axis=1)
imp.fit(sampl_x.loc[:,:])
nullRemoval = imp.transform(sampl_x.loc[:,:])
sample_x = pd.DataFrame(nullRemoval)
sample_x.index = sampl[0]
# sample_x.to_csv('../../yuChuli/tmp.csv')

imp = Imputer(missing_values='NaN', strategy='median', axis=1)
imp.fit(sampl_y.loc[:,:])
nullRemoval = imp.transform(sampl_y.loc[:,:])
sample_y = pd.DataFrame(nullRemoval)
```

```python
sample_x
```

```python
# 均值，中位数特征
# feature creation
sample_x_sum = sample_x.sum(axis=1)

sample_x_ratio_wk = (sample_x[[5,6]]).sum(axis=1)/(sample_x_sum.replace(0,1))

sample_x_mean = sample_x.mean(axis=1)
sample_x_median = sample_x.median(axis=1)
sample_x_std = sample_x.std(axis=1)

# 交叉特征
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2,interaction_only=True,include_bias=False)
jiaocha = poly.fit_transform(sample_x)
jiaochaFatures = pd.DataFrame(jiaocha)
jiaochaFatures['mean'] = sample_x_mean.values
jiaochaFatures['median'] = sample_x_median.values
jiaochaFatures['std'] = sample_x_std.values
jiaochaFatures['ratio_wk'] = sample_x_ratio_wk.values
jiaochaFatures['dates'] = sampl_dates
```

```python
flowSamples = pd.concat([jiaochaFatures,sample_y],axis=1)
```

```python
flowSamples.to_csv('../../models/RF/一次7天/10月份样本_3980_48.csv')
```
