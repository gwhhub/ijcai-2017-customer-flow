# 周末趋势

```python
# ds['store_id'].unique()

count = 0
index_without_weekTrend = []
for index in ds['store_id'].unique():
    
    tmp = ds[ds['store_id']==index]
    tmp.index = pd.to_datetime(tmp['date'])
    try:
#         tmp = tmp['2016-08':'2016-10']

        tmpWeekdays = tmp[ (tmp['weekday'] < 5 )]
        
        tmpWeekends = tmp[ (tmp['weekday'] == 6) | (tmp['weekday'] == 7)]
        
        tmpWeekdays_mean = tmpWeekdays['flow'].median(skipna=True)
        tmpWeekends_mean = tmpWeekends['flow'].median(skipna=True)

        
        if  ((tmpWeekdays_mean + 75) < tmpWeekends_mean) or ( (tmpWeekends_mean + 75) < tmpWeekdays_mean ):
            print (index ,end=' ')
            visualData(tmp, [index] ,'2016-08-01', '2016-10-31', '../overview/type/strongWeekends/',withTest=True)
            count = count + 1
        else:
            index_without_weekTrend.append(index)
    except:
        print ('???')
        continue
```

# 天气影响

```python
ds = pd.read_csv('../yuChuli/parsedWithTest.csv',index_col=0)
ds.index = pd.to_datetime(ds['date'])
```

```python
# ds['store_id'].unique()

count = 0
for index in ds['store_id'].unique():
    
    tmp = ds[ds['store_id']==index]
    tmp.index = pd.to_datetime(tmp['date'])
#     try:
#         tmp = tmp['2016-08':'2016-10']
    ordinaryDays = tmp[ (tmp['weather_1'] =='正常' )]
    rainDays = tmp[ (tmp['weather_1'] =='中雨' )]

    ordinaryDays_mean = ordinaryDays['flow'].median(skipna=True)
    rainDays_mean = rainDays['flow'].mean(skipna=True)


    if  ((rainDays_mean + 150) < ordinaryDays_mean) or ( (ordinaryDays_mean + 150) < rainDays_mean ):
        print (index ,end=' ')
        visualData(tmp, [index] ,'2016-08-01', '2016-10-31', '../overview/type/weathers/',withTest=True)
        count = count + 1
#     except:
#         print ('???')
#         continue
```
